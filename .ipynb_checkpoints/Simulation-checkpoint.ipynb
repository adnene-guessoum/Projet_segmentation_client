## K_means

```python
train_data.columns
```

```python
mO_data = train_data[['Recency', 'Frequency', 'Monetary', 'review_score', 'freight_value']]
mO_data.shape
```

```python
numeric_transformer = RobustScaler(unit_variance=True)
mO_scaled = numeric_transformer.fit_transform(mO_data)
mO_scaled
```

```python
model = cluster.KMeans()
visualizer = KElbowVisualizer(model, k=(1,10))

visualizer.fit(mO_scaled)    # Fit the data to the visualizer
visualizer.poof()    # Draw/show/poof the data
```

```python
# Instantiate the clustering model and visualizer 
model = cluster.KMeans(4)
visualizer = SilhouetteVisualizer(model)

visualizer.fit(mO_scaled)    # Fit the data to the visualizer
visualizer.poof()    # Draw/show/poof the data
```

```python
model.fit(mO_scaled)
```

```python
pred = model.predict(mO_scaled)
```

```python
print(model.labels_)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(mO_scaled[:,0], mO_scaled[:,1], mO_scaled[:,2], c=pred)
plt.show()
```

```python
from sklearn import metrics
metrics.adjusted_rand_score(pred, model.labels_)
```

```python
for i in range(len(test_groups)):
    mO_data = test_groups[i][['Recency', 'Frequency', 'Monetary', 'review_score', 'freight_value']]
    mO_scaled = numeric_transformer.fit_transform(mO_data)
    new_clust = cluster.KMeans(4).fit(mO_scaled)
    pred = model.predict(mO_scaled)
    print('month {} after first clustering :'.format(i+1), metrics.adjusted_rand_score(pred, new_clust.labels_))
    
```
